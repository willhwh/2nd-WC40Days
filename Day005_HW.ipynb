## 作業目標

* 請利用 API: https://www.dcard.tw/_api/forums/pet/posts?popular=true 回答下列問題：

1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

```python
import requests
r=requests.get("https://www.dcard.tw/_api/forums/pet/posts?popular=true")
response=r.text
```

```python
response
```

```python
import json
data = json.loads(response)
```

```python
r.json()
```

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
len(data)
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
for d in data:
    print('標題',d['title'])
    print('貼文時間',d['createdAt'])
    print('留言人數',d['commentCount'])
    print('按讚人數',d['likeCount'])

```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」
#熱門
hotcomment=0
hotlike=0
for d in data:
    hotcomment=hotcomment+d['commentCount']
    hotlike=hotlike+d['likeCount']
print('熱門平均留言人數 ',hotcomment/len(data))
print('熱門平均按讚人數 ',hotlike/len(data))
      
```

```python
#非熱門
r_not_hot=requests.get("https://www.dcard.tw/_api/forums/pet/posts?popular=false")
response_not_hot=r_not_hot.text
data_not_hot=json.loads(response_not_hot)
```

```python
len(data_not_hot)
```

```python
r_not_hot.json()
```

```python
for d in data_not_hot:
    print('標題',d['title'])
    print('貼文時間',d['createdAt'])
    print('留言人數',d['commentCount'])
    print('按讚人數',d['likeCount'])

```

```python
hotcomment1=0
hotlike1=0
for d in data_not_hot:
    hotcomment1=hotcomment1+d['commentCount']
    hotlike1=hotlike1+d['likeCount']
print('熱門平均留言人數 ',hotcomment1/len(data_not_hot))
print('熱門平均按讚人數 ',hotlike1/len(data_not_hot))
```
