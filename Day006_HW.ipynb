* 了解知乎 API 使用方式與回傳內容
* 撰寫程式存取 API 且添加標頭




```python
import requests
r = requests.get('https://www.zhihu.com/api/v4/questions/55493026/answers')
response = r.text

response
```

```python
headers = {'user-agent': 'my-app/0.0.1'}
r = requests.get('https://www.zhihu.com/api/v4/questions/55493026/answers',headers=headers)
response = r.text

response
```

```python
import json
data = json.loads(response)

for d in data['data']:
    print(d)

```

```python
r.json()
```

```python
def jprint(obj):
    # create a formatted string of the Python JSON object
    text = json.dumps(obj, sort_keys=True, 
                      indent=4)# tab 長度######################
    print(text)

jprint(r.json())
```

### 1. 取出知乎問題發問時間

```python
# 1. 取出知乎問題發問時間
for d in data['data']:
    print(d['created_time'])
```

### 2. 取出第一筆與最後一筆回答的時間

```python
# 2. 取出第一筆與最後一筆回答的時間
lst=[]
for d in data['data']:
    lst.append(d['created_time'])
print(lst[0],"\t",lst[-1])
```
