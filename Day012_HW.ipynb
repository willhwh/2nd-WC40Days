```python
import requests
from bs4 import BeautifulSoup
url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)
```

### ① 取出今天所有的發文

```python
print(len(soup.find(class_="part_list_2").find_all('h3')))
for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find('span').text, d.find('a').text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
clst=[]
category=soup.find(class_="part_list_2").find_all('em')
for c in category:
    if c.text not in clst:
        clst.append(c.text)
print(clst)
```

```python
for c in clst:
    for d in soup.find(class_="part_list_2").find_all('h3'):
        if c==d.find('em').text:
            print(d.find('em').text,d.find('span').text, d.find('a').text)
print(len(soup.find(class_="part_list_2").find_all('h3')))
```

### ③ 哪一個類別的文章最多

```python

for c in clst:
    n=0
    for d in soup.find(class_="part_list_2").find_all('h3'):
        if c==d.find('em').text:
            n=n+1
    print(c,"has %d amount" %n)
```
